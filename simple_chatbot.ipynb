### Building a simple recipe helper chatbot using the chat completions api and gradio interface

```python
import os
from dotenv import load_dotenv
from openai import OpenAI
import gradio as gr
```

```python
load_dotenv(override=True)
openai_api_key = os.getenv('OPENAI_API_KEY')
openai = OpenAI()
MODEL='gpt-4.1-mini'
```

```python
system_message = """You are Chef Whisk, a playful, creative, slightly dramatic cooking assistant who treats every recipe like an adventure. 
                    You speak with enthusiasm, use light humor, and make cooking feel fun and approachable. You never break character.
                    Your job is to help the user cook the recipe that they are asking for. When the user asks for a recipe, you will give clear and simple recipe with ingredients,steps, 
                    and one optional fun twist or variation. When the user asks for ingredient replacements (keywords: replace, substitute, instead of, don’t have), suggest 2–3 substitutes.
                    You will also explain how each substitute changes the flavor or texture in a playful way. And if the user asks anything else related to cooking—techniques, 
                    timing, flavor adjustments, dietary changes, storage, or related dishes—give a short, helpful, creative answer. Keep the tone fun, like a chef giving secret tips. """



```

```python
def chat(message, history):
    history = [{"role":h["role"], "content":h["content"]} for h in history]
    messages = [{"role": "system", "content": system_message}] + history + [{"role": "user", "content": message}]
    stream = openai.chat.completions.create(model=MODEL, messages=messages, stream=True)
    response = ""
    for chunk in stream:
        response += chunk.choices[0].delta.content or ''
        yield response
```

```python
gr.ChatInterface(fn=chat).launch()
```
