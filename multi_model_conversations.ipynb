#### A simple 3 way conversation model with different openai models (Alex - gpt-5-mini, Blake - gpt-5-nano and Charlie - gpt-4.1-mini).

```python
import os
from dotenv import load_dotenv
from openai import OpenAI
from IPython.display import Markdown, display
```

```python
load_dotenv(override=True)
openai_key = os.getenv("OPENAI_API_KEY")

openai=OpenAI()
```

```python
gpt_5_mini_system = "You are Alex. You are a chatbot who is very argumentative and responds in 1-2 sentences at maximum; \
you disagree with anything in the conversation and you challenge everything, in a brief and snarky way. \
    You are in a conversation with Blake and Charlie."

gpt_5_nano_system = "You are Blake. You are a very polite, courteous chatbot. You try to agree with \
everything the other person says, or find common ground. If the other person is argumentative, \
you try to calm them down and keep chatting. You try to respons with 2-3 sentences at maximum. \
    You are in a conversation with Alex and Charlie."

gpt_4_mini_system = "You are Charlie. You are a neutral chatbot. You neither agree nor disagree with anything the other person says. \
You will respond in a polite way as long as the other person is polite. If the other person is rude or argumentative, you will respond in a brief and curt way. \
    You are in a conversation with Blake and Alex."
```

```python

alex_user_prompt = """
You are Alex, in conversation with Blake and Charlie.
Which has the conversations that happend so far in the format:
"Alex": ....,
"Blake": ....,
"Charlie": ....", and so on.
Now with this, respond with what you would like to say next based on the conversation happened in this current execution frame, as Alex.
The conversation so far is as follows:
"""


blake_user_prompt = f"""
You are Blake, in conversation with Alex and Charlie.
Which has the conversations that happend so far in the format:
"Alex": ....,
"Blake": ....,
"Charlie": ....", and so on.
Now with this, respond with what you would like to say next based on the conversation happened in this current execution frame, as Blake.
The conversation so far is as follows:
""" 

charlie_user_prompt = """ You are Charlie, in conversation with Blake and Alex.
Which has the conversations that happend so far in the format:
"Alex": ....,
"Blake": ....,
"Charlie": ....", and so on.
Now with this, respond with what you would like to say next based on the conversation happened in this current execution frame, as Charlie.
The conversation so far is as follows:
"""
```

```python
convo_starter = input("Enter the conversation thread to start the conversation: ")
conversation = [{"Alex": convo_starter}]

def update_conversation():
    global all_conversation
    all_conversation = ''
    for convo in conversation:
        for key, value in convo.items():
            all_conversation += f' "{key}": "{value}"'

def call_gpt_5_mini():
    messages = [{"role": "system", "content": gpt_5_mini_system},{"role":"user","content":alex_user_prompt+all_conversation}]
    response = openai.chat.completions.create(model="gpt-5-mini", messages=messages)
    return response.choices[0].message.content

def call_gpt_5_nano():
    messages = [{"role": "system", "content": gpt_5_nano_system},{"role":"user","content":blake_user_prompt+all_conversation}]
    response = openai.chat.completions.create(model="gpt-5-nano", messages=messages)
    return response.choices[0].message.content

def call_gpt_4_mini():
    messages = [{"role": "system", "content": gpt_4_mini_system},{"role":"user","content":charlie_user_prompt+all_conversation}]
    response = openai.chat.completions.create(model="gpt-4.1-mini", messages=messages)
    return response.choices[0].message.content
```

```python
display(Markdown(f"### Alex:\n{conversation[0]['Alex']}\n"))

for i in range(5):
    # print(all_conversation)
    update_conversation()
    gpt_5_nano_next = call_gpt_5_nano()
    display(Markdown(f"### Blake:\n{gpt_5_nano_next}\n"))
    conversation.append({"Blake":gpt_5_nano_next})
    
    update_conversation()
    gpt_4_mini_next = call_gpt_4_mini()
    display(Markdown(f"### Charlie:\n{gpt_4_mini_next}\n"))
    conversation.append({"Charlie":gpt_4_mini_next})
    
    update_conversation()
    gpt_5_mini_next = call_gpt_5_mini()
    display(Markdown(f"### Alex:\n{gpt_5_mini_next}\n"))
    conversation.append({"Alex":gpt_5_mini_next})
    
    
    
    
```
